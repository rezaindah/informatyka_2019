```python
import matplotlib.pyplot as plt
import random

#https://machinelearningmastery.com/how-to-generate-random-numbers-in-python/
```

```python
x=[random.randint(1,10) for i in range(1000)]
#print(x)
plt.hist(x, bins=2)
plt.gca().set(title='Frequency histogram', ylabel='frequency')
plt.show
```

```python
# generate random Gaussian values
from random import seed
from random import gauss
# seed random number generator
seed(1)
# generate some Gaussian values
for _ in range(10):
    value = random.gauss(0,2)
    print(value)
    
plt.hist(value, bins=5)
plt.gca().set(title='Frequency histogram', ylabel='frequency')
plt.show
```

```python
import numpy as np
```

```python
our_list=[[1,2,3,4],[3,4,5,6],[5,6,7,8]]
array=np.array(our_list, dtype='float')

print(array)
print('Shape: ', array.shape)
print('Datatype: ',array.dtype)
print('Size: ', array.size)
print('Num Dimension: ', array.ndim)
```

```python
newlist=[[1,2,3],[4,5,6],[7,8,9]]
newarr=np.array(newlist, dtype='int')

print(newarr)
print('Shape: ', newarr.shape)
print('Datatype: ',newarr.dtype)
print('Size: ', newarr.size)
print('Num Dimension: ', newarr.ndim)
```

```python
print('Mean value is: ', newarr.mean())
print('Max value is: ', newarr.max())
print('Min value is: ', newarr.min())
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
iris = datasets.load_iris()
iris_data = iris.data
iris_labels = iris.target
```

```python
data1 = pd.DataFrame(data= np.c_[iris_data, iris_labels],
                     columns= iris['feature_names'] + ['target'])
data1.head()
```

```python
# menentukan training data dan testing data, gunakan random untuk indexing
np.random.seed(10)
# index data
indices = np.random.permutation(len(iris_data))
#print(indices)
# jumlah data yang di ambil
n_training_samples = 40
learnset_data = iris_data[indices[:-n_training_samples]]
learnset_labels = iris_labels[indices[:-n_training_samples]]
testset_data = iris_data[indices[-n_training_samples:]]
testset_labels = iris_labels[indices[-n_training_samples:]]
print(learnset_data, learnset_labels)
print(testset_data, testset_labels)
```

```python
# mencari jarak/similarity dengan euclidean
def distance(instance1, instance2):
    # buat jaga2 kalo instans nya adalah list atau tuple
    instance1 = np.array(instance1) 
    instance2 = np.array(instance2)
    
    return np.linalg.norm(instance1 - instance2)
print(distance([5.7, 3.8, 1.7, 0.3], [5,2.3,3.3,1]))
print(distance(learnset_data[87], learnset_data[9]))
```

```python
def get_neighbors(training_set, 
                  labels, 
                  test_instance, 
                  k, 
                  distance=distance):
    distances = []
    for index in range(len(training_set)):
        dist = distance(test_instance, training_set[index])
        distances.append((training_set[index], dist, labels[index]))
    distances.sort(key=lambda x: x[1])
    neighbors = distances[:k]
    return neighbors
```

```python
# jumlah testing data yg diambil hanya 3, dan melihat tetangganya siapa saja
for i in range(40):
    neighbors = get_neighbors(learnset_data, 
                              learnset_labels, 
                              testset_data[i], 
                              3, # jumlah neighbour yang bisa diubah
                              distance=distance)
    print(i, 
          "data testing=", testset_data[i], 
          "label data=",testset_labels[i],
          "Neighbours=" ,neighbors)
```

```python
#mencari kelas yg paling banyak muncul diantara neighbour
from collections import Counter
def claspred(neighbors):
    class_counter = Counter()
    for neighbor in neighbors:
        class_counter[neighbor[2]] += 1
    return class_counter.most_common(1)[0][0]
```

```python
numb = []
new_numb = 0
for i in range(n_training_samples):
    neighbors = get_neighbors(learnset_data, 
                              learnset_labels, 
                              testset_data[i], 
                              3, 
                              distance=distance)
    numb.append(claspred(neighbors))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(testset_labels, numb))
print(classification_report(testset_labels, numb))
```
